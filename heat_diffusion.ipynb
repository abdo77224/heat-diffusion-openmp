```python
serial_time = 11.7775
openmp_time = 3.8078
```

```python
import matplotlib.pyplot as plt

labels = ["Serial", "OpenMP"]
times = [serial_time, openmp_time]

plt.figure(figsize=(6,4))
plt.bar(labels, times)
plt.title("Heat Diffusion Performance Comparison")
plt.ylabel("Time (seconds)")

for i, v in enumerate(times):
    plt.text(i, v + 0.2, str(round(v, 2)), ha='center')

plt.show()
```

```python
speedup = serial_time / openmp_time
speedup
```

```python
plt.bar(["Speedup"], [serial_time / openmp_time])
plt.title("OpenMP Speedup")
plt.ylabel("Factor")
plt.show()
```
